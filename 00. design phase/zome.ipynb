```python
import numpy as np
# function (theta, m, n) {     return asin(cos(theta)^2 *sin(2*(PI rad )*m/n) / sqrt(1-(cos(theta)^2*cos(2*(PI rad)*m/n)+sin(theta)^2)^2)); }
# function (theta, m, n) {     return asin(cos(theta)^2 *sin(2*(PI rad )*m/n) / sqrt(1-(cos(theta)^2*cos(2*(PI rad)*m/n)+sin(theta)^2)^2)); }
def face_angle(theta,N,m):
    ''' Angle of a face m wrt main axis
        theta: Zome top angle
        N: Zome order
        m: Face number
        
        Zome top
        ----
        |\ Theta
        | \
        |Q \
    '''

    n = N
    alpha = np.asin(
        np.cos(theta)**2 *np.sin(2*np.pi*m/n) / 
        np.sqrt(1-(np.cos(theta)**2 * np.cos(2*np.pi*m/n) + np.sin(theta)**2)**2)
    )
    return alpha
### Dome parameters
Q = np.radians(49.51) ## Zome top angle (not theta)
L = 630.8 ## Rib length
N= 10

theta = np.pi/2 - Q 
face_angle(theta,N,1)
M1,M2=1,2
alpha = np.pi - (face_angle(theta, N,M1 ) - face_angle(theta,N,M2)) ## angle between planes (straight down)
np.degrees(alpha)
```

```python
def R_outer(N,L,Q):
    return L*np.sin(Q)/np.sin(2*np.pi/N/2)
    #L*sin(#Q)/sin(360deg/#N/2)

def angle_inner(M,N,L,R_outer):
    ''' L: Rib length
        R_outer: Outer radius of zome'''
    A = 2*np.pi/N
    return 2*np.asin(R_outer/L*np.sin(A/2)*np.sin(M*A/2))

theta_A1= angle_inner(1,N,L,R_outer(N,L,Q))

def gamma(theta,alpha):
    ''' Theta: inner angle of the parallellogram
        alpha: Angle between faces
        gamma: Angle...?
    '''
    return 2*np.atan(np.tan(theta/2)/np.sin(alpha/2))
np.degrees(gamma(theta_A1, alpha)),theta_A1

def q(theta, gamma):
    '''
        theta: inner angle parallelogram
        gamma: 
    '''
    return np.asin(np.sin(theta/2)/np.sin(gamma/2))

np.degrees(q(theta_A1,gamma(theta_A1,alpha)))
```

```python
from dataclasses import dataclass
## Point generator
@dataclass
class Zome():
    N : int
    L : float
    theta: float

    def faces(self,n=None, n2=None):
        if n is None:
            n = self.N
        if n2 is None:
            n2 = self.N
        faces = []
        for i in range(n):
            for j in range(i+1, i+n2):
                f = self.face(i,j)
                faces.append(f)
        return faces

    def face(self, i,j):
        ''' Collection of four vertices'''
        vs = []
        for oi in [0,1]:
            for oj in [0,-1]:
                v = self.v(i+oi,j+oj)
                vs.append(v)
        return vs

    def v(self, i,j):
        if i > j:
            return np.array([0,0,0])
        if i == j:
            return self.g(i)
        else:
            return np.sum([self.g(k) for k in range(i,j+1)],axis=0)

    def g(self, i):
        theta =self.theta
        N = self.N
        L = self.L
        
        # gi = (Cos[θ] Cos[360 i/n], Cos[θ] Sin[360 i/n], Sin[θ])
        x0 = np.cos(theta)*np.cos(np.pi*2*i/N)
        y0 = np.cos(theta)*np.sin(np.pi*2*i/N)
        z0 = np.sin(self.theta)
        # print(x0,y0,z0)
        return np.array([x0,y0,z0])

z = Zome(N,1,theta)

```

```python
import numpy as np
import pyvista as pv
def reorder_quad_vertices_ccw(face_pts):
    """
    Return quad vertices reordered consistently CCW and the face normal.
    """
    # Force float dtype to avoid in-place division errors
    pts = np.asarray(face_pts, dtype=float).reshape(4, 3)

    centroid = pts.mean(axis=0)

    # Best-fit plane normal from SVD
    U, S, Vt = np.linalg.svd(pts - centroid)
    normal = Vt[-1]
    normal /= np.linalg.norm(normal)

    # Build plane basis
    v0 = pts[0] - centroid
    if np.linalg.norm(v0) < 1e-12:
        v0 = pts[1] - centroid

    axis_x = v0 - normal * np.dot(v0, normal)
    if np.linalg.norm(axis_x) < 1e-12:
        axis_x = np.cross(normal, [1.0, 0.0, 0.0])
        if np.linalg.norm(axis_x) < 1e-12:
            axis_x = np.cross(normal, [0.0, 1.0, 0.0])

    axis_x = axis_x / np.linalg.norm(axis_x)
    axis_y = np.cross(normal, axis_x)
    axis_y = axis_y / np.linalg.norm(axis_y)

    # Compute angles for CCW sorting
    rel = pts - centroid
    xs = rel.dot(axis_x)
    ys = rel.dot(axis_y)
    angles = np.arctan2(ys, xs)

    order = np.argsort(angles)
    ordered = pts[order]

    # Ensure geometric normal aligns with SVD normal
    geo = np.cross(ordered[1] - ordered[0], ordered[2] - ordered[0])
    if np.linalg.norm(geo) > 1e-12:
        geo /= np.linalg.norm(geo)
        if np.dot(geo, normal) < 0:
            ordered = ordered[::-1]
            geo = -geo

    return ordered, normal


def plot_quad_mesh_from_face_vertices_ordered(faces, tol=1e-12, orient_outward=True):
    """
    faces: list of faces, each face is iterable of four [x,y,z] coordinates
    tol: tolerance for deduplication
    orient_outward: if True, ensure face normals point away from mesh centroid
    """

    # 1) Reorder vertices of each face to consistent CCW winding and collect face normals
    ordered_faces = []
    face_normals = []
    for face in faces:
        if len(face) != 4:
            raise ValueError("All faces must have exactly 4 vertices")
        ordered_pts, fn = reorder_quad_vertices_ccw(face)
        ordered_faces.append(ordered_pts)
        face_normals.append(fn)

    ordered_faces = np.array(ordered_faces)  # shape (F, 4, 3)
    face_normals = np.array(face_normals)    # shape (F, 3)

    # 2) Optionally orient normals outward w.r.t. mesh centroid
    mesh_centroid = ordered_faces.mean(axis=(0, 1))
    if orient_outward:
        for i in range(len(ordered_faces)):
            face_center = ordered_faces[i].mean(axis=0)
            to_center = face_center - mesh_centroid
            # If normal points towards the mesh centroid (i.e. into the mesh), flip face vertices
            if np.dot(face_normals[i], to_center) < 0:
                ordered_faces[i] = ordered_faces[i][::-1]
                face_normals[i] = -face_normals[i]

    # 3) Deduplicate vertices across faces (with tolerance)
    all_pts = ordered_faces.reshape(-1, 3)  # (F*4, 3)
    # Use rounding hash to merge near-equal points
    key = np.round(all_pts / tol).astype(np.int64)
    unique_map = {}
    unique_vertices = []
    index_list = []
    for k, real_p in zip(map(tuple, key), all_pts):
        if k not in unique_map:
            unique_map[k] = len(unique_vertices)
            unique_vertices.append(real_p)
        index_list.append(unique_map[k])
    unique_vertices = np.array(unique_vertices)

    # 4) Build faces as index lists in PyVista format
    quad_indices = np.array(index_list).reshape(-1, 4)  # (F, 4)
    pv_faces = np.hstack([[4, *q] for q in quad_indices])

    # 5) Build PyVista mesh and compute normals
    mesh = pv.PolyData(unique_vertices, pv_faces)
    mesh = mesh.compute_normals(auto_orient_normals=False)  # we already oriented faces
    # Plot
    plotter = pv.Plotter()
    plotter.add_mesh(mesh, show_edges=True, color="lightgray", smooth_shading=False)
    plotter.add_axes()
    plotter.enable_3_lights()
    plotter.show()

    return mesh  # return the PyVista mesh if caller wants to use it

```
